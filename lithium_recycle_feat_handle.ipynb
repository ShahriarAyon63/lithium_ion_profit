```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder


# Create a copy to avoid modifying the original DataFrame
df = data.copy()

# Handling null values by filling them with appropriate values or strategies
df.fillna(df.mean(), inplace=True)
df.fillna('Unknown', inplace=True)
df.dropna(axis=0, inplace=True)


# Now, you can proceed with label encoding or other preprocessing steps
label_encoder = LabelEncoder()

# Iterate through each column and apply label encoding to categorical columns
for column in ['Type', 'Cathode Scenario', 'Location scenario', 'location of recycling ', 'scenarios']:
    if df[column].dtype == 'O':
        df[column] = label_encoder.fit_transform(df[column])

# Save the processed DataFrame to a new Excel file
output_file_path = '/content/drive/MyDrive/lithium/OUTPUT_sensitivity33.csv'
df.to_csv(output_file_path, index=False)

# Print a message indicating the file has been saved
print(f"Processed data saved to: {output_file_path}")
```
